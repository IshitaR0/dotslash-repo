```python
import pandas as pd

# Assuming ishData is a pandas DataFrame and 'Price' is the column name

# Convert the 'Price' column to float, coercing errors to NaN
ishData['Price'] = pd.to_numeric(ishData['Price'], errors='coerce')

# Drop rows with NaN values in the 'Price' column
ishData = ishData.dropna(subset=['Price'])

# Sort ishData in descending order of price
ishData_sorted = ishData.sort_values(by='Price', ascending=False)

# Get the highest price
highest_price = ishData_sorted['Price'].max()

# Get the lowest price
lowest_price = ishData_sorted['Price'].min()
```

```python
# Display the sorted DataFrame, highest price, and lowest price
highest_price, lowest_price
```

```python
ishData_sorted.head()
```

```python
df_midhackuse = ishData_sorted[[
    "Name",
    "Description",
    "Brand_Name",
    "Category",
    "Subcategory",
    "Price",
    "CommonOrGenericNameOfCommodity",
]]

df_midhackuse.head()
```

```python
unique_categories = df_midhackuse['Category'].unique()
print(unique_categories)
```

```python
categories = df_midhackuse['Category'].unique()

for category in categories:
    subcategories = df_midhackuse[df_midhackuse['Category'] == category]['Subcategory'].unique()
    print(f"Category: {category}")
    print("Subcategories:")
    for subcategory in subcategories:
        print(f" - {subcategory}")
    print("\n")
```

```python
df_midhackuse.to_csv('df_midhackuse.csv', index=False)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(groceryData['Price'], bins=1000, edgecolor='black')
plt.title('Distribution of Prices in Grocery Data')
plt.xlabel('Price')
plt.xlim(0, 1000)
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
groceryData['Subcategory'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Distribution of Objects Over Subcategories in Grocery Data')
plt.xlabel('Subcategory')
plt.ylabel('Number of Objects')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
